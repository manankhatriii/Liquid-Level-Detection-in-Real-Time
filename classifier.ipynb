```python
import cv2
import numpy as np
import tensorflow_hub as hub
import tensorflow
import tf_keras
```

```python
model= tf_keras.models.load_model("H5/mobilenet_v2.h5", custom_objects={'KerasLayer': hub.KerasLayer}) 
```

```python
def preprocess_frame(frame):
    frame = cv2.resize(frame, (224, 224)) 
    frame = frame / 255.0
    frame = np.expand_dims(frame, axis=0)
    return frame
```

```python
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Error: Could not open video stream.")
else:
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        processed_frame = preprocess_frame(frame)
        prediction = model.predict(processed_frame)
        label_index = np.argmax(prediction[0])
        label_map = {0: 'Full', 1: 'Half', 2: 'Low', 3: 'Absent'}
        label = label_map[label_index]
        
        
        if label != 'Absent':
            cv2.putText(frame, label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        
        cv2.imshow("Pepsi Level Monitor", frame)
        cv2.resizeWindow("Pepsi Level Monitor", 250, 250)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
```
